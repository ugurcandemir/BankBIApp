```python
import pandas as pd
import numpy as np

# Load the first and second sheets of AKBNK.xlsx
sheet1 = pd.read_excel("AKBNK.xlsx", sheet_name=0)  # First sheet
sheet2 = pd.read_excel("AKBNK.xlsx", sheet_name=1)  # Second sheet

# Save the datasets to new Excel files
sheet1.to_excel("AKBNK_balance_1.xlsx", index=False)  # Save first sheet
sheet2.to_excel("AKBNK_income_1.xlsx", index=False)   # Save second sheet

print("Datasets saved as AKBNK_balance_1.xlsx and AKBNK_income_1.xlsx.")
```

```python
# Transform the "Kalem" column into column names
transformed_df = sheet1.set_index("Kalem").T.reset_index()

# Rename the columns
transformed_df.rename(columns={"index": "Yıllar"}, inplace=True)

# Display the transformed DataFrame
print(transformed_df.head())
```

```python
# Transform the "Kalem" column into column names for the first sheet
transformed_sheet1 = sheet1.set_index("Kalem").T.reset_index()
transformed_sheet1.rename(columns={"index": "Yıllar"}, inplace=True)
transformed_sheet1.fillna(0, inplace=True)  # Replace NaN with 0

# Transform the "Kalem" column into column names for the second sheet
transformed_sheet2 = sheet2.set_index("Kalem").T.reset_index()
transformed_sheet2.rename(columns={"index": "Yıllar"}, inplace=True)
transformed_sheet2.fillna(0, inplace=True)  # Replace NaN with 0

# Save the transformed datasets to new Excel files
transformed_sheet1.to_excel("AKBNK_balance_2.xlsx", index=False)  # Save first sheet
transformed_sheet2.to_excel("AKBNK_income_2.xlsx", index=False)   # Save second sheet

print("Transformed datasets saved as AKBNK_balance_2.xlsx and AKBNK_income_2.xlsx.")
```
